```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Input csv data

equipment_loss = pd.read_csv("../input/2022-ukraine-russian-war/russia_losses_equipment.csv", index_col='date', parse_dates=True)
personnel_loss = pd.read_csv("../input/2022-ukraine-russian-war/russia_losses_personnel.csv", index_col='date', parse_dates=True)

equipment_loss.head()
personnel_loss.head()
```

```python
#make copies of the datasets
equipment_loss_data = equipment_loss.copy()
personnel_loss_data = personnel_loss.copy()

personnel_loss_data.head()
```

```python
# Visualize the data
sns.set_style("darkgrid")
# Trends in damages over the past few months of war
plt.figure(figsize=(20,25))
plt.title("Trends in Number of Equipment destroyed in Russia-Ukraine War")
plt.ylabel("Equipment Losses")
equip_ = equipment_loss_data.drop(columns=['day'])
sns.lineplot(data=equip_)
```

```python
# trends in personnel loss

plt.figure(figsize=(20,25))
plt.title("Trends in Personnel Loss in Russia-Ukraine War")
plt.ylabel("Personnel Loss")
pers = personnel_loss_data.drop(columns=['day', 'personnel*'])
sns.lineplot(data=pers)
```

```python
# Visualize Equipment Loss by Category
# Air equipment which are Aircraft, Helicopter, Anti-aircraft and Drone

plt.figure(figsize=(20,25))
plt.title("Trend in Aircraft Equipment Losses")
plt.ylabel("Equipment Loss")
sns.lineplot(data=equip_['aircraft'], label='Aircraft')
sns.lineplot(data=equip_['helicopter'], label='Helicopter')
sns.lineplot(data=equip_['anti-aircraft warfare'], label='Anti-Aircraft Warfare')
sns.lineplot(data=equip_['drone'], label='Drone')
plt.legend()
```

```python
# other land weaponry and equipment loss
# Weaponry include Tanks, Field Artillery, APC, Military Auto
plt.figure(figsize=(20,25))
plt.title("Trend in Land Weaponry and Other Equipment Loss")
plt.ylabel("Equipment Loss")
sns.lineplot(data=equip_['tank'], label="Tanks")
sns.lineplot(data=equip_['field artillery'], label="Field Artillery")
sns.lineplot(data=equip_['APC'], label="APC")
sns.lineplot(data=equip_['military auto'], label="Military Auto")
plt.legend()
```

```python
# other weaponry loss
# Weaponry include Naval Ship, MRL, Special Equipment, Mobile SRBM, Fuel Tank

plt.figure(figsize=(20,25))
plt.title("Trend in Other Waeponry and Special Equipment Loss")
plt.ylabel("Equipment Loss")
sns.lineplot(data=equip_['naval ship'], label="Naval Ship")
sns.lineplot(data=equip_['MRL'], label="MRL")
sns.lineplot(data=equip_['special equipment'], label="Special Equipment")
sns.lineplot(data=equip_['mobile SRBM system'], label="Mobile SRBM")
sns.lineplot(data=equip_['fuel tank'], label="Fuel Tank")
plt.legend()
```
